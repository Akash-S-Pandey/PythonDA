```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv('Uber.csv')
df
```

```python
yields_apples = [0.895 , 0.91 ,0.919 ,0.926 , 0.929 , 0.931]
yields_orange = [0.89 , 0.91 , 0.919 , 0.921 , 0.918 , 0.914]
years = range(2011,2017)
years
```

```python
plt.figure(figsize=(11,5))
sns.set_style("darkgrid")

plt.plot(years,yields_apples,'s-r');
plt.plot(years,yields_orange,'o-b');

plt.xlabel('years');
plt.ylabel('yeild (ton per hectare)');

plt.title('fruit production in kanto');
plt.legend(["yields_apples","yields_orange"]);
```

sl like join in Pandas

```python
student1 = pd.DataFrame({
    "roll_no"     : [102,101,104,103,105],
    "name"        : ['Aravind','Rahul',"prateek",'piyuesh',"kartik"],
    "grade"       : ["B","B","A","C","A"],
    "marks"       : [15,15,20,4,22],
    "city"        : ['gurugram',"Delhi",'Delhi','gurugram','Hyderabad']


                         })
```

let create a mapping table

```python
city_state_mapping = pd.DataFrame(
    {
        "city"  : ["Gurugram","Delhi","Hydrabad","Faridabad","Allahabad"] ,
        "state" : ["Haryana","Delhi","Telangana","Haryana","Kanto"]
    }
)
```

```python
city_state_mapping
```

```python
final = student1.merge(city_state_mapping, how='left' , on = 'city' )
```

```python
final
```

```python
final.sort_values(by=['grade','marks'],ascending=[True,False])
```

```python
roll_no = pd.DataFrame({
    "roll_no" : [102,103]
})
roll_no
```

```python
final = student1.merge(roll_no, how='right' , on = 'roll_no' )
```

```python
final
```

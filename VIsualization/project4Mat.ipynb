```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline   
import seaborn as sns
```

```python
height = [150,160,165,185]
Weight = [70,80,90,100]
plt.plot(height,Weight);
 
sns.set_style('darkgrid')

plt.xlabel('Weight (kg)');
plt.ylabel('Height in (cm)');

plt.title("Relation between Height and weight");
```

```python
calorie_burn = [65,75,95,99]
Weight = [70,80,90,100]
plt.figure(figsize=(8,5))
plt.plot(calorie_burn,marker ='o',c='r')
plt.plot(Weight,marker ='s',c='g',ls='--')
sns.set_style('darkgrid')

plt.legend(['calorie_burn','Weight']);
plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4']);

plt.title('Relation between calorie and weight');

plt.xlabel('Weight in kg');
plt.ylabel("calorie_burn in (jules) ");
```

Creating thr multiple plots

```python
fig , ax = plt.subplots(nrows= 2, ncols=2 , figsize=(13,6),sharex=True,sharey=True)
sns.set_style('darkgrid');

ax[0,0].plot(calorie_burn,marker ='o',c='r');
ax[1,1].plot(Weight,marker ='s',c='b',ls='--');

ax[0,0].set_title('calorie_burn')
ax[1,1].set_title('Weight')

ax[0,0].set_xticks(ticks=[0,1,2,3])
ax[1,1].set_xticks(ticks=[0,1,2,3])


ax[0,0].set_xticklabels(labels=['p1','p2','p3','p4'])
ax[1,1].set_xticklabels(labels=['p1','p2','p3','p4'])






```

Visualiztion using Matplotlib

```python
data_BM = pd.read_csv('Data/big_mart_sales.csv')
data_BM = data_BM.dropna(how='any')
data_BM.head()
```

Line Chart

```python
price_by_item = data_BM.groupby('Item_Type').Item_MRP.mean()[:10]

x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

sns.set_style('darkgrid')
plt.figure(figsize=(19,8))

plt.title("Mean Price for Each Item  ")

plt.xlabel('Item_Type ')
plt.ylabel('Mean Price ')

plt.xticks(labels = x , ticks=np.arange(len(x)) );

plt.plot(x,y,marker = 'o',c='r');
```

Bar Chart

```python
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

sales_by_outlet_size.sort_values(ascending=True,inplace=True)

x =  sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

plt.xlabel("Outlet_Size")
plt.ylabel("Sales")

plt.title('Mean for the Outlet Size ')

plt.xticks(labels=x , ticks=np.arange(len(x)))

plt.bar(x  ,y ,color =['red','green','orange']);
```

Histogram

```python
#title
plt.title('Item MRP (price) distribution ');

plt.xlabel('Item MRP')
plt.ylabel('Freuency')

plt.hist(data_BM['Item_MRP'],bins=20 , color= 'lightblue');
```

```python
data = data_BM[['Item_Outlet_Sales']]

red_diamond = dict(markerfacecolor = 'r',marker='D')

plt.title("Item Sales distribution")

#make the box plot
plt.boxplot(data.values , labels=['Item_Sales'],flierprops=red_diamond);
```

```python
data = data_BM[['Item_Weight','Item_MRP']]

red_diamond = dict(markerfacecolor = 'r',marker='D')

fig , ax = plt.subplots()

#make the box plot
plt.boxplot(data.values , labels=['Item_Weight','Item_MRP'],flierprops=red_diamond);
```

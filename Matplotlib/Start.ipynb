```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Here Plotting the multiple Lines

Here as u can see the foll represents the production of apple (tons per hectare) over a six year 
in an imaginary country kanto

```python

yields_apples = [0.895 , 0.91 ,0.919 ,0.926 , 0.929 , 0.931]
yields_orange = [0.89 , 0.91 , 0.919 , 0.921 , 0.918 , 0.914]
years = range(2011,2017)
```

here can visualize the yeild production of the apples

```python
plt.plot(years,yields_apples); # here the plot(x,y) Represents the acctual graph
plt.plot(years,yields_orange)

plt.xlabel(' Years ');
plt.ylabel('Yield ( per hectare tons ) ');

plt.title(" Crop Yield in Kanto ");
plt.legend(["yields_apples","yields_orange"]);

```

```python
plt.plot(years,yields_apples , marker='o', c='b' , ls = '-' , lw=2 , ms = 8 , mew = 2 , mec = 'navy'); # marker can be represents
plt.plot(years,yields_orange , marker = "x" , c='r' , ls = '--',lw=3 , ms = 10 ,alpha= .5)

plt.xlabel(' Years ');
plt.ylabel('Yield ( per hectare tons ) ');

plt.title(" Crop Yield in Kanto ");
plt.legend(["yields_apples","yields_orange"]);

```

```python
plt.plot(years,yields_apples , 's-b'); # here the plot(x,y) Represents the acctual graph
plt.plot(years,yields_orange , 'o--r' )

plt.xlabel(' Years ');
plt.ylabel('Yield ( per hectare tons ) ');

plt.title(" Crop Yield in Kanto ");
plt.legend(["yields_apples","yields_orange"]);

```

```python
plt.figure(figsize=(11,5))  # use the figure size command to increse or decrease the figure size
plt.plot(years,yields_orange,'o--g');
plt.xlabel('Y e a r s');
plt.ylabel('Yield ( tons per hectare )');
plt.title("Y ield of Oranges in kanto");
```

```python
orignal_list = ["analytic","vidya",'is',"south ","east","king"]
filter_list = list(filter(lambda x : (x[0] == 's'),orignal_list))
print("the number starting with s is",filter_list)
```

```python
orignal_list = ["analytic","vidya",'is',"south ","east","king"]
filter_list = list(map(lambda x :( x[0].upper() + x[1:] ),orignal_list))
print("New List",filter_list)
```

```python
from functools import  reduce
orignal_list = [5,17,33,43,13,63,345,78,41]
filter_list = reduce((lambda x , y : x+y),orignal_list)
print("The sum of Number is : ",filter_list)
```

```python
from functools import  reduce
orignal_list = [5,17,33,43,13,63,345,78,41]
filter_list = reduce((lambda x,y : x if (x>y) else y),orignal_list)
print("The sum of Number is : ", filter_list)
```

```python
def count_hand_shakes(total_number_of_person):
    if total_number_of_person ==  2 :
        return 1
    else:
        return count_hand_shakes(total_number_of_person - 1) + (total_number_of_person-1)
```

```python
count_hand_shakes(5)
```

```python
def count_hand(total_number_of_person):
    count = 0
    for i in range(1,total_number_of_person):
        for j in range(i+1,total_number_of_person+1):
            print("Hand shake of ",i,"and",j)
            count = count + 1
    print("total no of hand hake",count)       

      
```

```python
count_hand(5)
```
